# 23. Reshaping a dataframe

```python
import pandas as pd

# Let's assume your DataFrame is called df
# df = pd.read_csv(...)  # or already loaded

melted_df = pd.melt(
    df,
    id_vars=['survived', 'sex', 'pclass'],
    value_vars=['age', 'fare', 'sibsp', 'parch'],
    var_name='Feature',
    value_name='Value'
)

melted_df.head()
```
