# Data Visualization in Python

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
df=sns.load_dataset('titanic')
df.head()
```

## Pandas plots

```python
df.plot(kind='scatter',x='age',y='fare',color='#bd0023')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar',color='red')
plt.title('Average Age of Males and Females')
plt.xticks(rotation=0)       
plt.ylabel('Average Fare')     
plt.xlabel('Sex')             
```

```python
plt.Figure(figsize=(15,13))
sns.scatterplot(df,x='age',y='fare',hue='sex')
```

```python
pal = {'male': 'red', 'female': 'blue'}

# Set theme
sns.set_theme(style='darkgrid', font_scale=1.2, font='Times New Roman')

# Create figure
plt.figure(figsize=(12, 8))

# Scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=pal)

# Labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Scatter plot of Age vs. Fare")
plt.legend(title='Gender')
# Show plot
plt.show()
```

```python
sns.barplot(df,x='sex',y=df.fare,hue='class',capsize=0.1)
```
